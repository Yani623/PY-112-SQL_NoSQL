## Reading in from a json file 

Let's read this json file of all meteorite landings on earth provided by NASA


````python
import urllib.request as url_portal, json 
JSON_URL = "https://data.nasa.gov/resource/y77d-th95.json"
with url_portal.urlopen(JSON_URL) as JSON_DATA:
    data = json.loads(JSON_DATA.read().decode())
    print(data)
````

## Let's create a mongo db out of this 

````python
from pymongo import MongoClient
from pprint import pprint
client = MongoClient()


#set this to True if you wanna populate mongodb 
populate_mongodb = False

try:
    if populate_mongodb:
        database = client[ "nasa-database" ]
        meteorite_collection = database[ "meteorite-landings" ]
        result=meteorite_collection.insert_many(data)
except Exception as e:
    raise e

# print(result.inserted_id)
# print(result.acknowledged)
````

### using limit 

````python
for entry in meteorite_collection.find().limit(4):
    print(entry)
````

### using the $or operator 

````python
res = meteorite_collection.find({
    "$or": [
      {"mass": {"$gte": 1000} },
      {"recclass": "H6" }
     ]
  })
for entry in res:
    print(entry)
````

### using the $and operator and get only the names


````python
res = meteorite_collection.find({
    "$or": [
      {"mass": {"$gte": 5000} },
      {"recclass": "L6" }
     ]
  })
for entry in res:
    print(entry['name'])
````

If you have used search engines like Google or Bing, you used the full-text search or FTS. The search engines collect contents from websites into databases and allow you to search based on keywords.

In addition to search engines, the FTS powers search results on websites like Blogs, News, E-commerce, and more.

Full-text search is a technique to search for documents that don’t perfectly match the search criteria. Documents are database entities that contain textual data e.g., product description, blog post, and articles.

## full text searching 


### using match/against
```

SELECT 
    productName, 
    productLine 
FROM products 
WHERE 
    MATCH(productLine) 
    AGAINST('Classic');
```
